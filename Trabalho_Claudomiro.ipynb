## Atividade 2.1

Criar um simulador de memória RAM seguindo o modelo dos slides. A memória possui 16 endereços, de `0000` até `1111`. Cada endereço armazena um dado de 8 bits. O programa deve permitir escrever, ler e listar os dados da memória.

<a id="principio"></a>
## Princípio de funcionamento

A memória RAM permite ler e escrever dados. Neste simulador, uma lista com 16 posições representa a memória. Todas as posições começam com `00000000`.

O endereço possui 4 bits. Como `2⁴ = 16`, ele pode selecionar uma das 16 posições. O dado possui 8 bits, ou seja, um byte. Na escrita, o dado é colocado na posição escolhida. Na leitura, o programa mostra o dado que está guardado naquela posição.

<a id="codigo"></a>
## Código do simulador

O código usa somente uma lista, um laço `while`, decisões com `if/elif/else` e um laço `for`. Para usar o simulador, execute a célula e digite as opções solicitadas. Qualquer tecla diferente de `W`, `R` ou `L` encerra o programa.

````python
# A memoria possui 16 posicoes, inicialmente zeradas.
memoria = ["00000000"] * 16

print("SIMULADOR DE MEMORIA RAM")
print("A memoria possui 16 enderecos e cada endereco guarda 8 bits.")

while True:
    print()
    print("Digite W para escrever.")
    print("Digite R para ler.")
    print("Digite L para listar toda a memoria.")
    print("Digite qualquer outra tecla para sair.")

    opcao = input("Escolha uma opcao: ").strip().upper()

    if opcao == "W":
        endereco = input("Digite um endereco de 4 bits: ").strip()

        if len(endereco) == 4 and endereco.count("0") + endereco.count("1") == 4:
            dado = input("Digite um dado de 8 bits: ").strip()

            if len(dado) == 8 and dado.count("0") + dado.count("1") == 8:
                posicao = int(endereco, 2)
                memoria[posicao] = dado
                print("Dado gravado com sucesso.")
            else:
                print("Dado invalido. Digite exatamente 8 bits.")
        else:
            print("Endereco invalido. Digite exatamente 4 bits.")

    elif opcao == "R":
        endereco = input("Digite um endereco de 4 bits: ").strip()

        if len(endereco) == 4 and endereco.count("0") + endereco.count("1") == 4:
            posicao = int(endereco, 2)
            print("Conteudo do endereco", endereco, ":", memoria[posicao])
        else:
            print("Endereco invalido. Digite exatamente 4 bits.")

    elif opcao == "L":
        print("Endereco : Conteudo")

        for posicao in range(16):
            endereco = format(posicao, "04b")
            print(endereco, ":", memoria[posicao])

    else:
        print("Simulador encerrado.")
        break
````

<a id="conclusao"></a>
## Conclusão

O simulador representa uma RAM com 16 endereços e palavras de 8 bits. Ele permite gravar um dado, ler um endereço e listar toda a memória. Os endereços e dados também são conferidos para impedir valores com quantidade errada de bits ou caracteres diferentes de `0` e `1`.

## Conceitos básicos (princípio de funcionamento)

Antes de resolver, deixo registrados os conceitos que vou usar o tempo todo. A memória principal é organizada em **células**, e o processador conversa com ela por meio de **barramentos** e de dois registradores (REM e RDM).

- **Célula:** menor unidade que tem endereço próprio na memória. Cada célula guarda uma quantidade fixa de bits.
- **Palavra (word):** unidade de informação que o processador manipula. Uma palavra pode caber em uma ou em mais de uma célula.
- **N (número de células / endereços):** com um barramento de endereços de **E** bits, existem **N = 2^E** endereços possíveis. Por isso o número de bits de endereço é **E = log₂(N)**.
- **Cuidado:** *quantidade de endereços* (N) é diferente de *largura do barramento de endereços* (E). N é quantos lugares existem; E é com quantos bits eu escrevo cada endereço.
- **REM (Registrador de Endereços de Memória):** guarda o endereço da célula a ser acessada. Tem o mesmo tamanho do barramento de endereços (**E** bits).
- **RDM (Registrador de Dados de Memória):** guarda o dado que entra ou sai da memória. Tem o tamanho do barramento de dados, ou seja, a quantidade de bits transferida por acesso.
- **Capacidade total (em bits)** = N × (bits por célula). Para passar para bytes, divido por 8.
- **Endereços vão de 0 até N − 1**, então o **maior endereço é N − 1**.

Conversões que vou usar (sempre em potências de 2):

| Unidade | Valor |
|---|---|
| 1 byte | 8 bits |
| 1 K | 2¹⁰ = 1.024 |
| 1 M | 2²⁰ = 1.048.576 |
| 1 G | 2³⁰ = 1.073.741.824 |

## Questão 1
*Classificação: RAM / memória principal — capacidade.*

> Um computador possui uma memória principal com capacidade para armazenar palavras de 16 bits em cada uma de suas N células, e o seu barramento de endereços tem 12 bits de tamanho. Sabendo-se que em cada célula pode-se armazenar o valor exato de uma palavra, quantos bytes poderão ser armazenados nessa memória?

**Princípio de funcionamento.** Cada célula guarda exatamente uma palavra de 16 bits, então cada célula tem 16 bits. O barramento de endereços tem 12 bits, e isso fixa o número de células: N = 2¹². A capacidade total em bits é N × 16, e depois passo para bytes dividindo por 8.

**Desenvolvimento.**

- N = 2¹² = 4.096 células
- Total de bits = 4.096 × 16 = 65.536 bits
- Total de bytes = 65.536 ÷ 8 = **8.192 bytes** (que é 8 KB)


````python
# Dados do enunciado
bits_endereco = 12        # o barramento de endereços tem 12 bits
bits_por_palavra = 16     # cada palavra tem 16 bits
# cada célula guarda 1 palavra, logo cada célula tem 16 bits

# Número de células: com 12 bits de endereço existem 2^12 endereços
numero_celulas = 2 ** bits_endereco
print("Número de células (2^12):", numero_celulas)

# Total de bits = número de células x bits por célula
total_bits = numero_celulas * bits_por_palavra
print("Total de bits:", total_bits)

# Converter bits para bytes (1 byte = 8 bits)
total_bytes = total_bits // 8
print("Total de bytes:", total_bytes)

# Conferindo em KB (1 KB = 1024 bytes)
print("Total em KB:", total_bytes / 1024)
````

**Conclusão.** A memória pode armazenar **8.192 bytes** (8 KB).

## Questão 5
*Classificação: RAM / memória principal — organização.*

> Qual é a diferença, em termos de endereço, conteúdo e total de bits, entre as seguintes organizações de MP:
> a) memória A: 32K células de 8 bits cada;
> b) memória B: 16K células de 16 bits cada; e
> c) memória C: 16K células de 8 bits cada?

**Princípio de funcionamento.** Para cada memória vou olhar três coisas: o **endereço** (quantos bits de endereço, que é log₂ do número de células), o **conteúdo** (bits por célula) e o **total de bits** (número de células × bits por célula). Uso K = 2¹⁰ = 1.024.

**Desenvolvimento.**

- Memória A: 32K = 32.768 = 2¹⁵ células → endereço de **15 bits**; conteúdo **8 bits**; total = 32.768 × 8 = **262.144 bits**.
- Memória B: 16K = 16.384 = 2¹⁴ células → endereço de **14 bits**; conteúdo **16 bits**; total = 16.384 × 16 = **262.144 bits**.
- Memória C: 16K = 16.384 = 2¹⁴ células → endereço de **14 bits**; conteúdo **8 bits**; total = 16.384 × 8 = **131.072 bits**.

Comparando: A precisa de 1 bit a mais de endereço do que B e C (porque tem mais células). A e C têm células de 8 bits, e B de 16 bits. O mais curioso é o total: **A e B têm exatamente o mesmo total de bits** (organizações diferentes, mesma capacidade), e **C tem metade** disso.


````python
# K = 1024 (2^10)
K = 1024

# Memória A: 32K células de 8 bits
celulas_A = 32 * K
bits_A = 8

# Memória B: 16K células de 16 bits
celulas_B = 16 * K
bits_B = 16

# Memória C: 16K células de 8 bits
celulas_C = 16 * K
bits_C = 8

# Bits de endereço da memória A: dobro 1 até chegar no nº de células e conto
bits = 0
valor = 1
while valor < celulas_A:
    valor = valor * 2
    bits = bits + 1
end_A = bits

# Bits de endereço da memória B
bits = 0
valor = 1
while valor < celulas_B:
    valor = valor * 2
    bits = bits + 1
end_B = bits

# Bits de endereço da memória C
bits = 0
valor = 1
while valor < celulas_C:
    valor = valor * 2
    bits = bits + 1
end_C = bits

# Total de bits de cada memória
total_A = celulas_A * bits_A
total_B = celulas_B * bits_B
total_C = celulas_C * bits_C

print("Memória A:", celulas_A, "células | endereço:", end_A, "bits | conteúdo:", bits_A, "bits | total:", total_A, "bits")
print("Memória B:", celulas_B, "células | endereço:", end_B, "bits | conteúdo:", bits_B, "bits | total:", total_B, "bits")
print("Memória C:", celulas_C, "células | endereço:", end_C, "bits | conteúdo:", bits_C, "bits | total:", total_C, "bits")
print()
print("A e B têm o mesmo total de bits?", total_A == total_B)
print("C tem metade do total de A?", total_C == total_A // 2)
````

**Conclusão.** Diferenças: **endereço** — A usa 15 bits, B e C usam 14 bits; **conteúdo** — A e C têm 8 bits por célula, B tem 16 bits; **total de bits** — A e B são iguais (262.144 bits) e C tem a metade (131.072 bits).

## Questão 10
*Classificação: RAM / memória principal.*

> Um computador possui um RDM com 16 bits de tamanho e um REM com capacidade para armazenar números com 20 bits. Sabe-se que a célula deste computador armazena dados com 8 bits de tamanho e que ele possui uma quantidade N de células, igual à sua capacidade máxima de armazenamento. Pergunta-se:
> a) Qual é o tamanho do barramento de endereços?
> b) Quantas células de memória são lidas em uma única operação de leitura?
> c) Quantos bits tem a memória principal?

**Princípio de funcionamento.** O REM tem o tamanho do barramento de endereços, então o item (a) sai direto. O RDM diz quantos bits são lidos por acesso; como cada célula tem 8 bits, o número de células por leitura é RDM ÷ tamanho da célula. Para o total, a capacidade máxima é N = 2²⁰ células (porque o REM tem 20 bits), e o total de bits é N × 8.

**Desenvolvimento.**

- a) Barramento de endereços = tamanho do REM = **20 bits**.
- b) Células por leitura = 16 ÷ 8 = **2 células**.
- c) N = 2²⁰ = 1.048.576 células; total = 1.048.576 × 8 = **8.388.608 bits** (que é 1 MB).


````python
# Dados do enunciado
rdm = 16          # RDM tem 16 bits
rem = 20          # REM guarda números (endereços) de 20 bits
bits_celula = 8   # cada célula guarda 8 bits
# N = capacidade máxima = 2^(bits do REM)

# a) Barramento de endereços = tamanho do REM
print("a) Barramento de endereços:", rem, "bits")

# b) Células lidas por leitura = bits do RDM / bits por célula
celulas_por_leitura = rdm // bits_celula
print("b) Células lidas por leitura:", celulas_por_leitura)

# c) Total de bits = N x bits por célula, com N = 2^20
numero_celulas = 2 ** rem
total_bits = numero_celulas * bits_celula
print("c) Número de células (2^20):", numero_celulas)
print("c) Total de bits da MP:", total_bits)
print("   Em bytes:", total_bits // 8)
print("   Em MB (1 MB = 1048576 bytes):", (total_bits // 8) / 1048576)
````

**Conclusão.** a) **20 bits**; b) **2 células** por leitura; c) **8.388.608 bits** (1 MB).

## Questão 11
*Classificação: RAM / memória principal.*

> Um microcomputador possui uma capacidade máxima de memória principal (RAM) com 32K células, cada uma capaz de armazenar uma palavra de 8 bits. Pergunta-se:
> a) Qual é o maior endereço, em decimal, desta memória?
> b) Qual é o tamanho do barramento de endereços deste sistema?
> c) Quantos bits podem ser armazenados no RDM e no REM?
> d) Qual é o total máximo de bits que pode existir nesta memória?

**Princípio de funcionamento.** São 32K = 2¹⁵ células, cada uma com 1 palavra de 8 bits. Os endereços vão de 0 até N − 1, então o maior endereço é N − 1. O barramento de endereços é log₂(N). O REM tem o tamanho do barramento de endereços e o RDM tem o tamanho da palavra (1 célula por acesso). O total é N × 8.

**Desenvolvimento.**

- a) Maior endereço = 32.768 − 1 = **32.767**.
- b) Barramento de endereços = log₂(32.768) = **15 bits**.
- c) REM = **15 bits**; RDM = **8 bits**.
- d) Total = 32.768 × 8 = **262.144 bits** (32 KB).


````python
# Dados do enunciado
K = 1024
numero_celulas = 32 * K     # 32K células
bits_palavra = 8            # cada célula guarda 1 palavra de 8 bits

# a) Maior endereço = N - 1 (endereços vão de 0 até N-1)
maior_endereco = numero_celulas - 1
print("a) Maior endereço (decimal):", maior_endereco)

# b) Barramento de endereços = log2(N): dobro 1 até chegar em N e conto
bits = 0
valor = 1
while valor < numero_celulas:
    valor = valor * 2
    bits = bits + 1
barramento_enderecos = bits
print("b) Barramento de endereços:", barramento_enderecos, "bits")

# c) REM = barramento de endereços ; RDM = tamanho da palavra (1 célula por acesso)
rem = barramento_enderecos
rdm = bits_palavra
print("c) REM:", rem, "bits | RDM:", rdm, "bits")

# d) Total de bits = N x bits por palavra
total_bits = numero_celulas * bits_palavra
print("d) Total máximo de bits:", total_bits)
print("   Em bytes:", total_bits // 8)
````

**Conclusão.** a) **32.767**; b) **15 bits**; c) REM = **15 bits** e RDM = **8 bits**; d) **262.144 bits** (32 KB).

## Questão 12
*Classificação: RAM / memória principal.*

> Considere uma célula de uma MP cujo endereço é, em hexadecimal, 2C81 e que tem armazenado em seu conteúdo um valor igual a, em hexadecimal, F5A. Sabe-se que, neste sistema, as células têm o mesmo tamanho das palavras e que em cada acesso é lido o valor de uma célula. Pergunta-se:
> a) Qual deve ser o tamanho do REM e do RDM nesse sistema?
> b) Qual deve ser a máxima quantidade de bits que podem ser implementados nessa memória?

**Princípio de funcionamento.** Cada dígito hexadecimal representa 4 bits. O endereço **2C81** tem 4 dígitos, então o REM precisa de pelo menos 4 × 4 = 16 bits. O conteúdo **F5A** tem 3 dígitos, então cada célula (que é igual à palavra) tem 3 × 4 = 12 bits, e como se lê uma célula por acesso, o RDM tem 12 bits. Para a quantidade máxima, suponho o REM com exatamente 16 bits, o que dá N = 2¹⁶ células.

*Hipótese:* o tamanho de cada registrador é dado pelo número de dígitos hexadecimais mostrados (16 bits para o endereço, 12 bits para o conteúdo).

**Desenvolvimento.**

- a) REM = 4 × 4 = **16 bits**; RDM = 3 × 4 = **12 bits**.
- b) N = 2¹⁶ = 65.536 células; máximo = 65.536 × 12 = **786.432 bits**.


````python
# Endereço e conteúdo dados em hexadecimal
endereco_hex = "2C81"
conteudo_hex = "F5A"

# Cada dígito hexadecimal vale 4 bits.
# Tamanho do REM = nº de dígitos do endereço x 4
rem = len(endereco_hex) * 4
# Tamanho do RDM = nº de dígitos do conteúdo x 4 (célula = palavra, 1 por acesso)
rdm = len(conteudo_hex) * 4
print("a) REM:", rem, "bits | RDM:", rdm, "bits")

# b) Máximo de bits = N x bits por célula, com N = 2^(bits do REM) e célula = RDM
numero_celulas = 2 ** rem
total_bits = numero_celulas * rdm
print("b) Número máximo de células (2^16):", numero_celulas)
print("b) Máximo de bits implementáveis:", total_bits)

# Só para conferir, mostro os valores em decimal
print("   Endereço 2C81 em decimal:", int(endereco_hex, 16))
print("   Conteúdo F5A em decimal:", int(conteudo_hex, 16))
````

**Conclusão.** a) REM = **16 bits** e RDM = **12 bits**; b) máximo de **786.432 bits**.

A partir do endereço (hexadecimal) 27FA, foram escritos sucessivamente grupos de 128 caracteres iguais, iniciando pelo grupo de As, seguido do grupo de Bs, e assim por diante. Qual deverá ser o endereço correspondente ao local onde está armazenado o 1.º J?

Os grupos têm 128 caracteres: o grupo de A começa em 27FA, o de B começa 128 células depois, o de C 256 depois, e assim por diante. A letra **J** é a **10ª** do alfabeto (A, B, C, D, E, F, G, H, I, J), então antes dela há **9 grupos completos** (de A até I). O endereço do 1.º J é o endereço inicial somado a 9 × 128.

## Questão 14
*Classificação: RAM / memória principal — tecnologia SRAM x DRAM.*

> O custo das memórias SRAM é maior que o das memórias DRAM. No entanto, o processo de conexão das memórias DRAM é mais complexo que o das SRAM e, em consequência, o preço da interface das DRAM é bem maior que o das SRAM. Supondo que uma interface de DRAM custe R\$ 5,00, uma interface de SRAM custe R\$ 1,00, o preço por bit de uma SRAM é de R\$ 0,00002 e o de uma DRAM é de R\$ 0,00001, calcule quantos bits deve ter uma memória dinâmica (DRAM) para que o conjunto seja mais barato.

**Princípio de funcionamento.** O custo total de cada tecnologia é o preço da interface mais o preço por bit vezes a quantidade de bits *n*. A DRAM tem interface mais cara (R\$ 5,00) mas custa menos por bit; a SRAM tem interface mais barata (R\$ 1,00) mas custa mais por bit. Existe um ponto de **empate**, e acima dele a DRAM compensa.

- Custo DRAM(n) = 5 + 0,00001 · n
- Custo SRAM(n) = 1 + 0,00002 · n

**Desenvolvimento.** Igualando os dois custos:

- 5 + 0,00001·n = 1 + 0,00002·n
- 5 − 1 = 0,00002·n − 0,00001·n
- 4 = 0,00001·n  →  n = 4 ÷ 0,00001 = **400.000 bits**

No empate, as duas custam R\$ 9,00. Para *n* **maior** que 400.000 bits, a DRAM fica mais barata.


````python
# Dados do enunciado (preços em reais)
interface_dram = 5.00
interface_sram = 1.00
preco_bit_sram = 0.00002
preco_bit_dram = 0.00001

# Custo total de cada tecnologia depende do nº de bits n:
# DRAM(n) = 5 + 0.00001 * n
# SRAM(n) = 1 + 0.00002 * n
# No empate: 5 + 0.00001*n = 1 + 0.00002*n  ->  4 = 0.00001*n  ->  n = 4 / 0.00001
n_empate = 4 / 0.00001
print("Ponto de empate (bits):", round(n_empate))

# Conferindo o custo das duas no ponto de empate
custo_dram = interface_dram + preco_bit_dram * n_empate
custo_sram = interface_sram + preco_bit_sram * n_empate
print("Custo DRAM no empate: R$", round(custo_dram, 2))
print("Custo SRAM no empate: R$", round(custo_sram, 2))

# Teste: com mais bits que o empate, a DRAM fica mais barata?
n_teste = n_empate + 100000
custo_dram2 = interface_dram + preco_bit_dram * n_teste
custo_sram2 = interface_sram + preco_bit_sram * n_teste
print()
print("Com", round(n_teste), "bits:")
print("Custo DRAM: R$", round(custo_dram2, 2))
print("Custo SRAM: R$", round(custo_sram2, 2))
print("DRAM mais barata?", custo_dram2 < custo_sram2)
````

**Conclusão.** O empate ocorre em **400.000 bits** (as duas custam R\$ 9,00). A DRAM só fica mais barata quando a memória tem **mais de 400.000 bits**.

*Classificação: RAM / memória principal — endereçamento.*

> Quantos bits são requeridos para se endereçar células em uma memória de 128G?

**Princípio de funcionamento.** Para endereçar N células preciso de E bits tais que 2^E ≥ N. Aqui N = 128G células.

*Hipótese:* G = Giga = 2³⁰ (potência de 2).

**Desenvolvimento.**

- 128G = 128 × 2³⁰ = 2⁷ × 2³⁰ = 2³⁷
- Logo, são necessários **37 bits**.


````python
# Memória de 128G células. Quero os bits de endereço.
# G = 2^30 (Giga) e 128 = 2^7, então 128G = 2^7 * 2^30 = 2^37.
G = 2 ** 30
numero_celulas = 128 * G
print("Número de células:", numero_celulas)

# Conto quantas vezes dobro 1 até chegar no nº de células.
bits = 0
valor = 1
while valor < numero_celulas:
    valor = valor * 2
    bits = bits + 1
print("Bits de endereço necessários:", bits)
````

**Conclusão.** São necessários **37 bits** para endereçar uma memória de 128G células.

## Questão 31
*Classificação: RAM / memória principal — capacidade.*

> Uma imagem pode ser representada por uma matriz de pontos armazenada na memória de um computador. Cada ponto possui uma indicação de cor associada a ela; essa cor precisa de 4 bytes para ser representada. Baseado nessas informações pede-se:
> a) a quantidade de memória, em bytes, necessária para armazenar uma imagem de 640 × 420 pontos;
> b) a quantidade de memória em megabytes necessária para armazenar 10 imagens semelhantes a esta;
> c) quantas imagens como esta poderiam ser armazenadas na memória de um computador com 128MB de memória RAM?

**Princípio de funcionamento.** O número de pontos é largura × altura. Cada ponto ocupa 4 bytes, então os bytes de uma imagem são (pontos × 4). Para 10 imagens, multiplico por 10 e passo para megabytes. Para saber quantas cabem em 128 MB, divido a memória pelo tamanho de uma imagem e fico só com a parte inteira (não dá para guardar imagem pela metade).

*Hipótese:* 1 MB = 2²⁰ bytes = 1.048.576 bytes (potência de 2).

**Desenvolvimento.**

- a) Pontos = 640 × 420 = 268.800; bytes = 268.800 × 4 = **1.075.200 bytes**.
- b) 10 imagens = 10.752.000 bytes ÷ 1.048.576 ≈ **10,25 MB**.
- c) 128 MB = 134.217.728 bytes; 134.217.728 ÷ 1.075.200 ≈ 124,8 → **124 imagens** completas.


````python
# Dados do enunciado
largura = 640        # pontos na horizontal
altura = 420         # pontos na vertical
bytes_por_ponto = 4  # cada cor ocupa 4 bytes

# a) Bytes para uma imagem = nº de pontos x bytes por ponto
pontos = largura * altura
bytes_imagem = pontos * bytes_por_ponto
print("a) Pontos por imagem:", pontos)
print("a) Bytes por imagem:", bytes_imagem)

# b) 10 imagens, resultado em megabytes (1 MB = 1048576 bytes)
bytes_10 = 10 * bytes_imagem
mb_10 = bytes_10 / 1048576
print("b) Bytes para 10 imagens:", bytes_10)
print("b) Em megabytes:", round(mb_10, 2), "MB")

# c) Quantas imagens completas cabem em 128 MB de RAM
ram_bytes = 128 * 1048576
quantas = ram_bytes // bytes_imagem   # divisão inteira: só cabem imagens completas
print("c) RAM em bytes:", ram_bytes)
print("c) Quantidade de imagens completas:", quantas)
````

**Conclusão.** a) **1.075.200 bytes** por imagem; b) ≈ **10,25 MB** para 10 imagens; c) cabem **124 imagens** completas em 128 MB de RAM.

<a id="principio"></a>
## 2. Princípio de funcionamento da ROM

**ROM** vem do inglês *Read Only Memory*, ou seja, **memória somente de leitura**. Suas principais características são:

- O conteúdo é **gravado previamente** e, **durante o uso normal, só pode ser lido** — não é possível escrever nele.
- É uma memória **não volátil**: o conteúdo permanece guardado mesmo sem energia.

Isso a diferencia da memória **RAM**, que permite tanto leitura quanto escrita durante o funcionamento.

Na ROM da Figura 4.18, ler um endereço significa **selecionar uma das 4 palavras gravadas** e colocar os seus 4 bits nas saídas **S₃ S₂ S₁ S₀**. Quem faz essa seleção é o decodificador, explicado a seguir.

<a id="decodificador"></a>
## 3. Decodificador 2 para 4 e o sinal de habilitação

**Decodificador 2 para 4:** recebe os **2 bits de endereço** (A₁ e A₀) e ativa **exatamente uma** de suas **4 linhas de saída** (00, 01, 10 e 11). Como são 2 entradas, existem 2² = **4 combinações possíveis**, e cada combinação corresponde a uma das 4 células da ROM.

A linha ativada é escolhida pelo valor do endereço, somando os pesos dos bits:

- A₁ é o bit da esquerda e tem **peso 2**;
- A₀ é o bit da direita e tem **peso 1**.

Por exemplo, o endereço **10** significa A₁ = 1 e A₀ = 0, que seleciona a linha 1×2 + 0×1 = **linha 2** (a palavra 1110). A linha ativada conduz os bits dessa palavra até as saídas S₃ S₂ S₁ S₀ por meio das portas OR mostradas na figura.

**Sinal de habilitação (*Habilitar*):** funciona como uma chave geral da ROM.

- Com **Habilitar = 1**, o decodificador fica ativo e a palavra selecionada aparece nas saídas.
- Com **Habilitar = 0**, nenhuma linha do decodificador é ativada. Como a figura usa portas OR e não mostra buffers tri-state, este modelo considera as saídas em **0000**.

## 5. Código do simulador

A ideia já foi explicada acima; agora vem o código. Para ficar fácil de entender, a ROM é guardada em **duas listas que andam juntas**: uma com os endereços e outra com os conteúdos. A leitura é feita **passo a passo**, do mesmo jeito que o circuito faria: conferir o endereço, ativar a linha do decodificador, ler a palavra e mostrar as saídas. O endereço a consultar fica guardado em uma **variável**, assim o notebook roda do início ao fim sem travar e sempre mostra o mesmo resultado.

````python
# A memoria ROM do slide 77 tem 4 celulas e cada celula guarda 4 bits.
# Como ainda estou aprendendo, vou guardar a ROM em duas listas simples.
# A lista "enderecos" tem os 4 enderecos possiveis (2 bits cada).
# A lista "conteudos" tem a palavra de 4 bits guardada em cada endereco.
# As duas listas andam juntas: a posicao 0 de uma combina com a posicao 0 da outra.

enderecos = ["00", "01", "10", "11"]
conteudos = ["0101", "1011", "1110", "0000"]

# Esta ROM e SOMENTE LEITURA: o conteudo acima foi gravado antes
# e nao pode ser alterado enquanto a memoria esta sendo usada.

# Vou mostrar a ROM para conferir se ficou igual a tabela do slide 77.
print("Conteudo gravado na ROM (somente leitura):")
for posicao in range(4):
    print("Endereco " + enderecos[posicao] + "  ->  Conteudo " + conteudos[posicao])
````

### Leitura de um endereço (passo a passo)

````python
# Aqui eu faco a LEITURA de um endereco da ROM, passo a passo,
# do mesmo jeito que o circuito do slide 77 faria.

# 1) O endereco que eu quero ler.
#    Deixo o endereco numa variavel para o notebook rodar inteiro sem travar.
#    Para testar outro endereco, e so trocar o valor abaixo (ex.: "00", "01", "10", "11").
endereco = "10"

# 2) O sinal de habilitacao (Habilitar): 1 = ROM ligada, 0 = ROM desligada.
habilitar = 1

# 3) Antes de ler, confiro se o endereco tem exatamente 2 bits
#    e se cada caractere e 0 ou 1.
endereco_valido = True
if len(endereco) != 2:
    endereco_valido = False
for bit in endereco:
    if bit != "0" and bit != "1":
        endereco_valido = False

# 4) Agora eu uso o resultado da conferencia.
if endereco_valido == False:
    print("Endereco invalido:", endereco)
    print("O endereco precisa ter exatamente 2 bits, usando so 0 e 1 (ex.: 00, 01, 10, 11).")
else:
    # O decodificador 2 para 4 olha os dois bits do endereco.
    # A1 e o bit da esquerda (peso 2) e A0 e o bit da direita (peso 1).
    A1 = endereco[0]
    A0 = endereco[1]

    # Monto as 4 linhas do decodificador comecando todas em 0.
    # So UMA delas vai virar 1 (a linha escolhida pelo endereco).
    linhas = [0, 0, 0, 0]

    # Descubro qual linha ligar somando os pesos dos bits.
    posicao = 0
    if A1 == "1":
        posicao = posicao + 2
    if A0 == "1":
        posicao = posicao + 1

    if habilitar == 1:
        # Com a ROM habilitada, o decodificador liga a linha escolhida.
        linhas[posicao] = 1

        # A linha ligada seleciona a palavra guardada naquele endereco.
        palavra = conteudos[posicao]

        # As saidas S3 S2 S1 S0 sao os 4 bits dessa palavra.
        S3 = palavra[0]
        S2 = palavra[1]
        S1 = palavra[2]
        S0 = palavra[3]

        print("Habilitar = 1 (ROM ligada)")
        print("Endereco lido:", endereco, "( A1 =", A1, "/ A0 =", A0, ")")
        print("Linhas do decodificador (00 01 10 11):", linhas)
        print("Linha ativada: linha", posicao)
        print("Palavra lida:", palavra)
        print("Saidas -> S3 =", S3, " S2 =", S2, " S1 =", S1, " S0 =", S0)
    else:
        # Com a ROM desligada, nenhuma linha e ativada.
        print("Habilitar = 0 (ROM desligada)")
        print("Nenhuma linha do decodificador e ativada.")
        print("Saidas -> S3 = 0  S2 = 0  S1 = 0  S0 = 0")
````

## Princípio de funcionamento

Um sistema de memória pode usar vários chips:

- Chips em **paralelo** aumentam a largura do barramento de dados.
- **Bancos** aumentam a capacidade sem aumentar a largura do BD.
- Com multiplexação **RAS/CAS**, os pinos recebem primeiro a linha e depois a coluna. Assim, 13 pinos físicos podem representar 26 bits internos.
- O **barramento lógico de endereços** inclui os bits internos e os bits usados para escolher o banco.

Adoto `G = 2³⁰`. Os projetos abaixo são soluções possíveis, pois os slides permitem mais de uma organização.

<a id="projeto1"></a>
## Projeto 1 - Slide 19

> Projetar uma memória com 13 pinos de endereço e capacidade de 8 Gbits. O BD transporta 2 palavras, não pode ultrapassar 64 bits e a palavra deve ser maior que a célula.

### Escolhas do projeto

1. **Palavra = 32 bits.** Como o BD transporta duas palavras, `2 × 32 = 64 bits`, que é o limite permitido.
2. **Célula = 16 bits.** A palavra fica maior que a célula e cada palavra é formada por duas células.
3. **Multiplexação RAS/CAS.** Os 13 pinos são usados em duas etapas, formando 26 bits internos.
4. **Bancos.** São usados para completar a capacidade de 8 Gbits.

### Cálculo passo a passo

**1. Capacidade solicitada**

`8 Gbits = 8 × 2³⁰ = 2³³ bits`.

**2. Capacidade de um chip**

Com 26 bits internos, cada chip possui `2²⁶` posições. Cada posição guarda uma célula de 16 bits:

`2²⁶ × 16 = 2³⁰ bits = 1 Gbit`.

Portanto, cada chip é organizado como **64 M × 16**.

**3. Formação do BD**

Cada chip fornece 16 bits. Para formar o BD de 64 bits:

`64 ÷ 16 = 4 chips em paralelo`.

Um banco possui 4 chips e capacidade de:

`4 × 1 Gbit = 4 Gbits`.

**4. Quantidade de bancos**

`8 Gbits ÷ 4 Gbits = 2 bancos`.

Logo, o sistema usa `2 × 4 = 8 chips`. Um bit escolhe entre os dois bancos.

**5. Barramento de endereços**

`26 bits internos + 1 bit de banco = 27 bits`.

Como o BD transfere duas palavras juntas, cada endereço lógico seleciona um grupo de 64 bits.

````python
# Verificacao do Projeto 1
G = 2 ** 30
palavra = 32
celula = 16
BD = 2 * palavra
pinos = 13
bits_internos = pinos * 2
capacidade_chip = (2 ** bits_internos) * celula
chips_por_banco = BD // celula
capacidade_banco = capacidade_chip * chips_por_banco
bancos = (8 * G) // capacidade_banco
total_chips = chips_por_banco * bancos
barramento_enderecos = bits_internos + 1
capacidade_total = capacidade_banco * bancos

print("Palavra =", palavra, "bits")
print("Celula =", celula, "bits")
print("BD =", BD, "bits")
print("Palavra maior que celula?", palavra > celula)
print("Chips por banco =", chips_por_banco)
print("Bancos =", bancos)
print("Total de chips =", total_chips)
print("Barramento de enderecos =", barramento_enderecos, "bits")
print("Capacidade correta?", capacidade_total == 8 * G)
````

`2²⁶ posições × 16 bits = 2³⁰ bits = 1 Gbit`.

Cada chip é um **64 M × 16**.

**3. Quantidade de bancos**

Como `32 = 2⁵`, são necessários 5 bits para selecionar um banco.

**4. Conferência do endereço**

O sistema usa 32 chips em profundidade. Apenas um banco fica ativo em cada acesso, mantendo o BD com 16 bits.

## Item 1 — Chip de memória 4.096K × 1 (4 Mbits)

*Figura analisada: figura da **direita (b)** do slide 5 de "Memória Tannembaun.pptx.pdf". A figura da esquerda (512K × 8) não é o objeto desta análise.*

> **INSERIR DESENHO DO SLIDE 5 AQUI**

**Enunciado.** Analisar o chip 4.096K × 1 (4 Mbits), que possui os pinos de endereço A0 a A10 e os sinais RAS, CAS, D, CS, WE e OE. Explicar a organização do chip, a quantidade de endereços, a largura dos dados, a capacidade total, a função dos sinais e como 11 pinos multiplexados conseguem representar 22 bits de endereço.

### Princípio de funcionamento

Este chip é uma memória dinâmica (DRAM). As células de 1 bit ficam organizadas internamente como uma **matriz de linhas e colunas**. Para gastar menos pinos, o endereço é **multiplexado no tempo**: as mesmas linhas A0–A10 transportam primeiro o **endereço de linha** e depois o **endereço de coluna**.

O acesso a uma célula acontece em duas etapas:

1. O endereço de linha é colocado em A0–A10 e travado pelo sinal **RAS** (escolhe a linha da matriz).
2. O endereço de coluna é colocado nas mesmas linhas A0–A10 e travado pelo sinal **CAS** (escolhe a coluna da matriz).

O cruzamento da linha com a coluna seleciona **uma** célula. Como existe apenas o pino **D** de dados, cada leitura ou escrita movimenta **1 bit**. O sinal **CS** habilita o chip, **WE** define se a operação é de escrita e **OE** habilita a saída na leitura.

### Função dos sinais

- **A0–A10:** 11 linhas de endereço, usadas de forma multiplexada (primeiro linha, depois coluna).
- **RAS (Row Address Strobe):** trava/valida o endereço de **linha**. Ativo em nível baixo.
- **CAS (Column Address Strobe):** trava/valida o endereço de **coluna**. Ativo em nível baixo.
- **D:** único pino de dados, bidirecional; transporta **1 bit** (entra na escrita e sai na leitura).
- **CS (Chip Select):** seleciona o chip; ele só responde quando este sinal está ativo.
- **WE (Write Enable):** quando ativo, a operação é de **escrita**; caso contrário, é de **leitura**.
- **OE (Output Enable):** habilita o buffer de **saída** dos dados durante a leitura.

### Desenvolvimento passo a passo

Convenções: 1K = 1024 e 1M = 1024 × 1024 (potências de 2). No slide, "4.096 K" usa o ponto como separador de milhar, ou seja, 4.096K = 4096K.

1. **Organização:** matriz de células, com palavra de **1 bit**, endereçada por linha (RAS) e coluna (CAS).
2. **Quantidade de endereços:** 4096 × 1024 = **4 194 304** = 4M = 2²² posições.
3. **Largura dos dados:** **1 bit** (apenas o pino D).
4. **Capacidade total:** 4 194 304 × 1 = **4 194 304 bits = 4 Mbits**.
5. **Bits de endereço:** 2²² = 4 194 304, logo são necessários **22 bits**.

### Como 11 pinos multiplexados representam 22 bits de endereço

Os 11 pinos A0–A10 são **reaproveitados em dois momentos**:

- 1º momento: o chip recebe **11 bits de linha** (travados por RAS);
- 2º momento: o chip recebe **11 bits de coluna** (travados por CAS).

Somando: 11 (linha) + 11 (coluna) = **22 bits lógicos** de endereço, que correspondem a 2²² = 4M células. Assim, com apenas 11 pinos físicos o chip consegue endereçar 22 bits — é essa a vantagem da multiplexação por RAS/CAS.

````python
# Item 1 - chip 4.096K x 1 (4 Mbits)
# Aqui eu so confiro a capacidade usando contas simples.

# 1K de memoria vale 1024 e 1M vale 1024 * 1024 (potencias de 2)
um_K = 1024
um_M = 1024 * 1024

# o slide escreve "4.096 K"; o ponto e separador de milhar, entao 4.096K = 4096K
numero_de_enderecos = 4096 * um_K   # quantas posicoes o chip tem
largura_dos_dados = 1               # o chip tem so o pino D, ou seja 1 bit

# capacidade total em bits = numero de enderecos vezes a largura
capacidade_bits = numero_de_enderecos * largura_dos_dados

print("Numero de enderecos:", numero_de_enderecos)
print("Confere que e igual a 4M?", numero_de_enderecos == 4 * um_M)
print("Largura dos dados (bits):", largura_dos_dados)
print("Capacidade total (bits):", capacidade_bits)
print("Capacidade em Mbits:", capacidade_bits / um_M)

# agora eu conto quantos bits de endereco sao necessarios.
# vou dobrando o numero 1 ate alcancar a quantidade de enderecos.
bits_de_endereco = 0
valor = 1
while valor < numero_de_enderecos:
    valor = valor * 2
    bits_de_endereco = bits_de_endereco + 1

print("Bits de endereco necessarios:", bits_de_endereco)

# como o endereco e multiplexado, ele entra em duas partes (linha e coluna)
bits_da_linha = int(bits_de_endereco / 2)
bits_da_coluna = int(bits_de_endereco / 2)
print("Bits de linha (travados por RAS):", bits_da_linha)
print("Bits de coluna (travados por CAS):", bits_da_coluna)
````

### Conclusão do Item 1

O chip 4.096K × 1 é uma DRAM com palavra de 1 bit e 4M posições, totalizando **4 Mbits**. Seus 11 pinos de endereço (A0–A10) são multiplexados por RAS e CAS, representando **22 bits** lógicos (11 de linha + 11 de coluna). O código confirmou a capacidade de 4 Mbits e os 22 bits de endereço.

Explicar a organização e a capacidade, a largura dos dados, a multiplexação de linha e coluna, a função dos bancos e dos sinais, e por que 128M posições exigem 27 bits lógicos de endereço.

### Princípio de funcionamento

Este chip também é uma DRAM e possui **4 bancos internos**. Os dois pinos Banco 0 e Banco 1 formam um número de 2 bits, capaz de selecionar um banco entre quatro.

O acesso ocorre assim:

1. Banco 0/Banco 1 escolhem o banco interno.
2. A0-A12 recebem o endereço de linha, guardado por RAS.
3. Os mesmos pinos recebem o endereço de coluna, guardado por CAS.
4. A posição escolhida transfere 4 bits por D0-D3.

CS seleciona o chip, WE controla a escrita e OE habilita a saída na leitura.

### Função dos sinais e dos bancos

- **A0-A12:** 13 pinos usados em momentos diferentes para linha e coluna.
- **RAS:** guarda o endereço de linha.
- **CAS:** guarda o endereço de coluna.
- **Banco 0/Banco 1:** 2 bits de seleção, portanto `2² = 4 bancos`.
- **D0-D3:** barramento de dados de 4 bits.
- **CS:** seleciona o chip.
- **WE:** habilita a escrita.
- **OE:** habilita a saída na leitura.

Os sinais aparecem com barra na figura, indicando que são ativos em nível baixo.

### Multiplexação e os 27 bits de endereço

`128M = 2²⁷`, portanto são necessários **27 bits lógicos** para escolher uma posição.

Os dois pinos de banco fornecem 2 bits. Restam:

`27 - 2 = 25 bits` para linha e coluna.

Uma organização possível usa **13 bits de linha + 12 bits de coluna**. Como linha e coluna passam em momentos diferentes pelos mesmos pinos, apenas 13 pinos físicos A0-A12 são necessários. Na fase de 12 bits, um dos pinos não é usado.

Conferindo: `13 + 12 + 2 = 27 bits`.

### Desenvolvimento passo a passo

1. **Organização:** 128M posições, cada uma com 4 bits.
2. **Posições:** `128 × 2²⁰ = 2²⁷`.
3. **Largura dos dados:** 4 bits, pelos pinos D0-D3.
4. **Capacidade:** `2²⁷ × 4 = 2²⁹ bits = 512 Mbits`.
5. **Endereço:** 13 bits de linha + 12 de coluna + 2 de banco = 27 bits.
6. **Bancos:** Banco 0/Banco 1 selecionam `2² = 4 bancos`.

````python
# Item 2 - chip 128M x 4 (512 Mbits)
um_M = 1024 * 1024
numero_de_posicoes = 128 * um_M
largura_dos_dados = 4
capacidade_bits = numero_de_posicoes * largura_dos_dados

# 128M = 2^27, logo sao necessarios 27 bits logicos.
bits_de_endereco = 27
bits_de_banco = 2
numero_de_bancos = 2 ** bits_de_banco
bits_de_linha = 13
bits_de_coluna = bits_de_endereco - bits_de_banco - bits_de_linha

print("Numero de posicoes:", numero_de_posicoes)
print("Largura dos dados:", largura_dos_dados, "bits")
print("Capacidade em Mbits:", capacidade_bits / um_M)
print("Bits logicos de endereco:", bits_de_endereco)
print("Bancos:", numero_de_bancos)
print("Linha + coluna + banco =", bits_de_linha, "+", bits_de_coluna, "+", bits_de_banco)
print("Total correto?", bits_de_linha + bits_de_coluna + bits_de_banco == bits_de_endereco)
````

### Conclusão do Item 2

O chip possui 128M posições de 4 bits, totalizando **512 Mbits**. Os 27 bits lógicos podem ser organizados em 13 bits de linha, 12 de coluna e 2 de seleção de banco. Os dois pinos Banco 0/Banco 1 selecionam **4 bancos**. O código confirmou a capacidade e a divisão do endereço.
