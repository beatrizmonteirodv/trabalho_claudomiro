## Atividade 2.1

Criar um simulador de memória RAM seguindo o modelo dos slides. A memória possui 16 endereços, de `0000` até `1111`. Cada endereço armazena um dado de 8 bits. O programa deve permitir escrever, ler e listar os dados da memória.

<a id="principio"></a>
## Princípio de funcionamento

A memória RAM permite ler e escrever dados. Neste simulador, uma lista com 16 posições representa a memória. Todas as posições começam com `00000000`.

O endereço possui 4 bits. Como `2⁴ = 16`, ele pode selecionar uma das 16 posições. O dado possui 8 bits, ou seja, um byte. Na escrita, o dado é colocado na posição escolhida. Na leitura, o programa mostra o dado que está guardado naquela posição.

<a id="codigo"></a>
## Código do simulador

O código usa somente uma lista, um laço `while`, decisões com `if/elif/else` e um laço `for`. Para usar o simulador, execute a célula e digite as opções solicitadas. Qualquer tecla diferente de `W`, `R` ou `L` encerra o programa.

```python
# A memoria possui 16 posicoes, inicialmente zeradas.
memoria = ["00000000"] * 16

print("SIMULADOR DE MEMORIA RAM")
print("A memoria possui 16 enderecos e cada endereco guarda 8 bits.")

while True:
    print()
    print("Digite W para escrever.")
    print("Digite R para ler.")
    print("Digite L para listar toda a memoria.")
    print("Digite qualquer outra tecla para sair.")

    opcao = input("Escolha uma opcao: ").strip().upper()

    if opcao == "W":
        endereco = input("Digite um endereco de 4 bits: ").strip()

        if len(endereco) == 4 and endereco.count("0") + endereco.count("1") == 4:
            dado = input("Digite um dado de 8 bits: ").strip()

            if len(dado) == 8 and dado.count("0") + dado.count("1") == 8:
                posicao = int(endereco, 2)
                memoria[posicao] = dado
                print("Dado gravado com sucesso.")
            else:
                print("Dado invalido. Digite exatamente 8 bits.")
        else:
            print("Endereco invalido. Digite exatamente 4 bits.")

    elif opcao == "R":
        endereco = input("Digite um endereco de 4 bits: ").strip()

        if len(endereco) == 4 and endereco.count("0") + endereco.count("1") == 4:
            posicao = int(endereco, 2)
            print("Conteudo do endereco", endereco, ":", memoria[posicao])
        else:
            print("Endereco invalido. Digite exatamente 4 bits.")

    elif opcao == "L":
        print("Endereco : Conteudo")

        for posicao in range(16):
            endereco = format(posicao, "04b")
            print(endereco, ":", memoria[posicao])

    else:
        print("Simulador encerrado.")
        break
```

<a id="conclusao"></a>
## Conclusão

O simulador representa uma RAM com 16 endereços e palavras de 8 bits. Ele permite gravar um dado, ler um endereço e listar toda a memória. Os endereços e dados também são conferidos para impedir valores com quantidade errada de bits ou caracteres diferentes de `0` e `1`.
