## Princípio de funcionamento

Chips em **paralelo** aumentam a largura do barramento de dados. **Bancos** aumentam a capacidade. Na multiplexação RAS/CAS, os mesmos pinos recebem primeiro a linha e depois a coluna. Assim, 13 pinos físicos representam 26 bits internos.

Neste projeto, `G = 2³⁰`. Pinos físicos e barramento lógico não são a mesma coisa.

## Projeto 1 - Slide 19

Projetar uma memória de **8 Gbits**, com **13 pinos de endereço** e BD que transporte **2 palavras**, sem ultrapassar 64 bits. A palavra deve ser maior que a célula.

### Solução

Adoto multiplexação RAS/CAS, palavra de **32 bits** e célula de **16 bits**.

- BD = 2 × 32 = **64 bits**.
- 13 pinos multiplexados = **26 bits internos**.
- Um chip: 2²⁶ posições × 16 bits = **1 Gbit** (`64 M × 16`).
- 4 chips em paralelo formam 64 bits e um banco de **4 Gbits**.
- 2 bancos formam **8 Gbits**, totalizando **8 chips**.
- Um bit escolhe o banco: barramento lógico = 26 + 1 = **27 bits**.

Cada endereço lógico seleciona o grupo de 64 bits, isto é, as duas palavras transferidas juntas.

| Grandeza | Valor |
|---|---:|
| Palavra | 32 bits |
| Célula | 16 bits |
| BD | 64 bits |
| Pinos físicos | 13 |
| Barramento lógico | 27 bits |
| Organização | 2 bancos de 4 chips |
| Capacidade | 8 Gbits |

```python
# Verificacao do Projeto 1
G = 2 ** 30
palavra = 32
celula = 16
BD = 2 * palavra
bits_internos = 13 * 2
capacidade_chip = (2 ** bits_internos) * celula
chips_por_banco = BD // celula
capacidade_banco = capacidade_chip * chips_por_banco
bancos = (8 * G) // capacidade_banco
barramento_enderecos = bits_internos + 1
capacidade_total = capacidade_banco * bancos

print("BD =", BD, "bits")
print("Palavra maior que celula?", palavra > celula)
print("Chips por banco =", chips_por_banco)
print("Bancos =", bancos)
print("Barramento de enderecos =", barramento_enderecos, "bits")
print("Capacidade correta?", capacidade_total == 8 * G)
```

## Projeto 2 - Slide 20

Projetar uma memória de **32 Gbits**, com **16 linhas de dados** e menos de 14 pinos físicos de endereço.

### Solução

Adoto 13 pinos multiplexados e células de 16 bits, iguais ao BD.

- 32 Gbits ÷ 16 bits = **2³¹ posições**.
- O barramento lógico precisa de **31 bits**.
- 13 pinos multiplexados representam **26 bits internos**.
- Um chip `64 M × 16` possui **1 Gbit**.
- São necessários **32 bancos**, com um chip em cada banco.
- 5 bits selecionam um dos 32 bancos: 26 + 5 = **31 bits**.

| Grandeza | Valor |
|---|---:|
| BD | 16 bits |
| Pinos físicos | 13 (< 14) |
| Barramento lógico | 31 bits |
| Organização | 32 bancos de 1 chip |
| Capacidade | 32 Gbits |

```python
# Verificacao do Projeto 2
G = 2 ** 30
capacidade = 32 * G
BD = 16
posicoes = capacidade // BD
barramento_logico = 31
pinos_fisicos = 13
bits_internos = pinos_fisicos * 2
capacidade_chip = (2 ** bits_internos) * BD
bancos = capacidade // capacidade_chip
bits_de_banco = 5
capacidade_total = capacidade_chip * bancos

print("Pinos menores que 14?", pinos_fisicos < 14)
print("Numero de posicoes = 2^31?", posicoes == 2 ** 31)
print("Bancos =", bancos)
print("Barramento de enderecos =", bits_internos + bits_de_banco, "bits")
print("Capacidade correta?", capacidade_total == 32 * G)
```
